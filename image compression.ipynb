```python
'''import every library needed for the algorithm to run'''

from scipy import misc
import numpy as np
from PIL import Image
import matplotlib.pyplot as plt
from scipy.ndimage import zoom
```

```python
'''function to show a picture'''

def show_image(img, title=None, grayscale=True):
    if grayscale:
        plt.imshow(img, cmap=plt.cm.gray)
    else:
        plt.imshow(img)
    plt.title(title)
    plt.show()
```

```python
'''import a jpg, show the original and transform it to a grayscale image, and then to an array'''

def image_preprocess(image_path):
    #open the image file
    img = Image.open(image_path)

    #show the original picture
    show_image(img, title='Original picture', grayscale=False)
    
    #convert the image to grayscale if it's not already
    img_gray = img.convert('L')
    
    #convert the image to a numpy array
    img_array = np.array(img_gray)

    #show the grayscale image
    show_image(img_gray, title='Grayscale image')

    return img_array
```

```python
'''function to compress the image by combining the given n neighbouring pixels into one'''

def compress_image (arr: np.ndarray, n: int) -> np.ndarray:
    #compression itself, explain it a bit
    result = np.zeros(shape=(int(arr.shape[0] / n), int(arr.shape[1] / n)))

    for idx, x in np.ndenumerate(result):
        lst = arr[idx[0] * n: (idx[0] + 1) * n, idx[1] * n: (idx[1] + 1) * n]
        result[idx] = np.median(lst)

    show_image(result, title='Compressed image')

    return result
```

```python
def reconstruct_original_size(compressed: np.ndarray, original_size: tuple[int, int]):
    return zoom(compressed, (original_size[0] / compressed.shape[0], original_size[1] / compressed.shape[1]))
```

```python
def test_compressions (original_image: np.ndarray, max_compression_ratio: int):

    result: dict[int, int] = {}

    for compression_ratio in range(2, max_compression_ratio):
        compressed = compress_image(original_image, compression_ratio)
        resized_compressed = reconstruct_original_size(compressed, original_image.shape)
        mse = np.square(resized_compressed - original_image).mean()
        print(f"{mse} for {compression_ratio}")
        result[compressed.size] = mse

    return result
```
